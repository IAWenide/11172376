```python
mean_chl = redWine_data['chlorides'].mean()
mean_fa = redWine_data['fixed acidity'].mean()
mean_fsd = redWine_data['free sulfur dioxide'].mean()
mean_dty = redWine_data['density'].mean()
mean_ql = redWine_data['quality'].mean()
mean_va = redWine_data['volatile acidity'].mean()
```

```python
print(f'chlorides: {mean_chl}')
print(f'fixed acidity: {mean_fa}')
print(f'free sulfur dioxide: {mean_fsd}')
print(f'density: {mean_dty}')
print(f'quality: {mean_ql}')
print(f'volatile acidity: {mean_va}')
```

```python
redWine_data['chlorides'] = redWine_data['chlorides'].fillna(mean_chl)

redWine_data['fixed acidity'] = redWine_data['fixed acidity'].fillna(mean_fa)
redWine_data['free sulfur dioxide'] = redWine_data['free sulfur dioxide'].fillna(mean_fsd)
redWine_data['density'] = redWine_data['density'].fillna(mean_dty)
redWine_data['quality'] = redWine_data['quality'].fillna(mean_ql)
redWine_data['volatile acidity'] = redWine_data['volatile acidity'].fillna(mean_va)
```

```python
redWine_data.groupby('quality').count().plot()
mpl.title('Plot Diagram representing the quality of the red wine')
mpl.savefig('graph1.png')
```

```python
redWine_data.plot(x = 'pH', y = 'density', xlabel = 'pH value', ylabel = 'density', kind = 'scatter')
mpl.title('A Scatter Plot of Density against pH value')
mpl.savefig('graph2')
```

```python
fixed_acidity = redWine_data.groupby(redWine_data['fixed acidity'].head(60))
total_fixed_acidity = fixed_acidity['fixed acidity'].agg('sum')
total_fixed_acidity.plot(kind = 'bar', title = 'A bar graph of Fixed Acidity of the red wine')
mpl.savefig('graph3')
```

```python
residual_sugar = redWine_data.groupby(redWine_data['residual sugar'].head(50))
total_residual_sugar = residual_sugar['residual sugar'].agg('sum')
total_residual_sugar.plot(kind = 'bar', title = 'A bar graph of Residual Sugar of the red wine')
mpl.savefig('graph4')
```
